```
import pandas_datareader.data as web
all_data = {ticker:web.get_data_yahoo(ticker) for ticker in ['CL=F','BZ=F','GC=F']}
import pandas as pd
price=pd.DataFrame({ticker:data['Adj Close'] for ticker,data in all_data.items()})
volume=pd.DataFrame({ticker:data['Volume'] for ticker,data in all_data.items()})
returns=price.pct_change()
print('Price percentage Change')
print('*************************')
print(returns.tail())
print('\nCORRELATION')
print('***********')
print(returns.corr())
print('\nCOVARIANCE')
print('*************')

print(returns.cov())

#print(returns.corrwith(returns['CL=F']))
print('\n')
print(price.describe())
#print(volume.describe())
import seaborn as sbs
sbs.pairplot(returns)
```
